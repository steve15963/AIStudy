# 데이터 세트 로드

Data 150개의 iris꽃의 특성을 보여줌.
    Numpy구성
        [0] 꽃 받침 길이
        [1] 꽃 받침 너비
        [2] 꽃 잎 길이
        [3] 꽃 잎 너비

```python
from sklearn import datasets

iris = datasets.load_iris()
data = iris.data
labels = iris.target

print(f"data : {data}")
```

# 다층 퍼셉트론(MLP) 로드

```python
from sklearn.neural_network._multilayer_perceptron import MLPClassifier
mlp = MLPClassifier(random_state=1)
print(f"mlp: {mlp}")
```

# 학습용 테스트용 데이터셋 분리

```python
from sklearn.model_selection import train_test_split

# test_size의 비율만큼 ( 0.5 ) 학습용 테스트 데이터, 라벨로 나눕니다.
data_train, data_test, labels_train, labels_test = train_test_split(data, labels, test_size= 0.5, random_state=1)
```

# 데이터 표준 분포 스케일 변경

```python
from sklearn.preprocessing import StandardScaler

#표준편차 Scaler 로드
scaler = StandardScaler()

# data 학습
scaler.fit(data_train)

print(data_train)
data_train_std = scaler.transform(data_train)
print(data_train_std)
data_test_std = scaler.transform(data_test)
```

# 다층 퍼셉트론 학습

```python
#학습
mlp.fit(data_train,labels_train)
```

# 다측 퍼셉트론 추론

```python
#예측
pred = mlp.predict(data_test)
```

# 다층 퍼셉트론 결과 확인

```python
from sklearn.metrics import accuracy_score

#계산
diff = labels_test != pred

print(f"잘 못 분류된 예시 {(labels_test != pred).sum()}")

print(f"정확도 : {accuracy_score(labels_test,pred)}")
```

# 그래픽 설정

```python
import numpy
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt

data = data[:,[0,2]]

markers = ('s','*','^')

colors= ('green','blue','red')

cmap = ListedColormap(colors)

x_min, x_max = data[:,0].min() - 1, data[:,0].max() + 1
y_min, y_max = data[:,1].min() - 1, data[:,1].max() + 1
resoultion = 0.01
x, y = numpy.meshgrid(
    numpy.arange(x_min,x_max,resoultion),
    numpy.arange(y_min,y_max,resoultion)
)

from sklearn.neural_network._multilayer_perceptron import MLPClassifier
mlp = MLPClassifier(random_state=1,max_iter=200,hidden_layer_sizes=(200,100))

mlp.fit(data,labels)

z = mlp.predict(
    numpy.array(
        [x.ravel(),y.ravel()]
    ).T
)

z = z.reshape(x.shape)

plt.pcolormesh(x,y,z,cmap=cmap)
plt.xlim(x.min(),x.max())
plt.ylim(y.min(),y.max())


classes = ["setosa", "versicolor", "verginica"]

for index,cl in enumerate(numpy.unique(labels)):
    plt.scatter(data[labels == cl, 0], data[labels == cl, 1], color=cmap(index), marker=markers[index],s=60,label=classes[index],edgecolors='black')

plt.xlabel("꽃잎 길이")
plt.ylabel("꽃 받침 길이")

plt.legend(loc='upper left')

plt.show()
```
